**Постоение графика зависимости кросс-корреляции от сдвига**

```python
from scipy.interpolate import interp1d
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cross_cor_FP1_plus.txt", sep="\t", header=None)
x = np.array(df[0])
y = np.array(df[1])

df1 = pd.read_csv("cross_cor_FP1_minus.txt", sep="\t", header=None)

x1 = np.array(df1[0])
y1 = np.array(df1[1])

f = interp1d(x, y)
f2 = interp1d(x, y, kind ='cubic')

f1 = interp1d(x1, y1)
f3 = interp1d(x1, y1, kind ='cubic')

plt.plot(x, f2(x), '-', x1, f3(x1), '--')

plt.legend(["Прямая цепь", "Обратная цепь"])
plt.xlabel('Сдвиг', size = 15)
plt.ylabel("Кросс-корркляция", size=15)
plt.show()
```
